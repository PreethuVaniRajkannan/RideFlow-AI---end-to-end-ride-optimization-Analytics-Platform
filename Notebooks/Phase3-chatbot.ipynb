```python
import pandas as pd
from googletrans import Translator

# =========================
# LOAD DATA
# =========================
data = pd.read_csv(r"C:\Users\chaka\Preethu\My_Git_Repo\Final Project_Rideflow\Datasets\preprocessed_rideflow_datasets.csv")

# Take one ride (example)
ride = data.iloc[0]

translator = Translator()

# =========================
# INTENT DETECTION
# =========================
def detect_intent(text):
    text = text.lower()

    if "where" in text or "driver" in text:
        return "track_driver"
    elif "eta" in text or "arrive" in text:
        return "eta"
    elif "cancel" in text:
        return "cancel"
    elif "price" in text or "fare" in text:
        return "price"
    else:
        return "general"

# =========================
# RESPONSE GENERATOR
# =========================
def generate_response(intent, ride):
    if intent == "track_driver":
        return f"Your driver is on the way and will arrive in {ride['estimated_eta_min']} minutes."
    
    elif intent == "eta":
        return f"Estimated arrival time is {ride['estimated_eta_min']} minutes."
    
    elif intent == "cancel":
        return "Your ride can be cancelled from the app. Cancellation charges may apply."
    
    elif intent == "price":
        return f"Your estimated fare is based on demand and distance."
    
    else:
        return "How can I assist you with your ride?"

# =========================
# TRANSLATION
# =========================
def auto_translate(text):
    detected = translator.detect(text).lang
    return translator.translate(text, dest=detected).text

# =========================
# CHATBOT LOOP
# =========================
print("🤖 AI Ride Chatbot (type 'exit' to stop)")
print("Supported languages: en, hi, ta\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Bot: Goodbye 👋")
        break

    lang = input("Preferred language (en/hi/ta): ").lower()

    intent = detect_intent(user_input)
    response = generate_response(intent, ride)

    translated = translate_text(response, lang)

    print("Bot:", translated)
```
